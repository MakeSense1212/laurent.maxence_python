```python
import random

class Joueur:
    
    def __init__(self, nom):
        self.__nom = nom
        
    def show(self):
        print(self.nom)
       
    
class carte(Joueur):
    def setValeur(self, valeur):
        self.valeur = valeur
    
    def setCouleur(self, couleur):
        self.couleur = couleur
        
class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    
    
    
def distribution(nom):
    nbCarte = 0

    while (nbCarte < 7):

        ran = random.random()

        if(ran <= 0.2):
            if(cartesBleues):
                carte = random.choice(cartesBleues)
                cartesBleues.remove(carte)
                nom.setValeur(carte)
                nom.setcouleur("Bleu")
                print(bcolors.OKBLUE + carte)
            else:
                nbCarte -=1

        if(ran > 0.2 and ran <= 0.4):
            if(cartesVertes):
                carte = random.choice(cartesVertes)
                cartesVertes.remove(carte)
                nom.setValeur(carte)
                nom.setcouleur("Vert")
                print(bcolors.OKGREEN + carte)
            else:
                nbCarte -=1

        if(ran > 0.4 and ran <= 0.6):
            if(cartesRouges):
                carte = random.choice(cartesRouges)
                cartesRouges.remove(carte)
                nom.setValeur(carte)
                nom.setcouleur("Rouge")
                print(bcolors.FAIL + carte)
            else:
                nbCarte -=1

        if(ran > 0.6 and ran <= 0.8):
            if(cartesJaunes):
                carte = random.choice(cartesJaunes)
                cartesJaunes.remove(carte)
                nom.setValeur(carte)
                nom.setcouleur("Jaune")
                print(bcolors.WARNING + carte)
            else:
                nbCarte -=1

        if(ran > 0.8 and ran <= 1):
            if(cartesJoker):
                carte = random.choice(cartesJoker)
                cartesJoker.remove(carte)
                nom.setValeur(carte)
                nom.setcouleur("aucune")
                print(bcolors.ENDC + carte)
            else:
                nbCarte -=1

        nbCarte += 1
        

        
cartesBleues = ['reverse','reverse','+2','+2','9','9','8','8','7','7','6','6','5','5','4','4','3','3','2','2','1','1','0']
cartesVertes = ['reverse','reverse','+2','+2','9','9','8','8','7','7','6','6','5','5','4','4','3','3','2','2','1','1','0']
cartesRouges = ['reverse','reverse','+2','+2','9','9','8','8','7','7','6','6','5','5','4','4','3','3','2','2','1','1','0']
cartesJaunes = ['reverse','reverse','+2','+2','9','9','8','8','7','7','6','6','5','5','4','4','3','3','2','2','1','1','0']
cartesJoker = ['+4','+4','+4','+4','couleur','couleur','couleur','couleur']
    
print("Bienvenu dans le jeu de UNO")

ok = "false"
while (ok == "false"):
    nbjoueur = input("Combien de joueur voulez-vous affronter ?")
    if(int(nbjoueur) == 0 ):
        print("Veuillez saisir un plus grand nombre de joueur Ã  affronter")
    elif(int(nbjoueur) > 9 ):
        print("Vous ne pouvez pas affronter plus de 9 joueurs")
    else:
        print("La partie commence, voici vos cartes")
        ok = "true"

count = 0
while (count < int(nbjoueur)):
    print (bcolors.HEADER + 'Distribution des cartes pour le joueur', count)
    distribution(count)
    count += 1     


print("La carte au dessus du packet est : ")
print("A vous de jouez, voici vos cartes : ")










```
